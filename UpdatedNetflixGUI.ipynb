```python
import pymysql
from tkinter import *

class netflixInterface:
    def __init__(self, master):
        
        #Set Master
        self.master = master
        
        #Set Title 
        master.title("CSE 305 - Netflix Database")
        
        #Set Window Width*Heigth
        master.geometry("500x500")
        
        #Set Database Connection
        self.connection = pymysql.connect("localhost", "root", "Diallo1!", "Netflix")
        self.cur = self.connection.cursor()
        
        #Sets up GUI 
        self._setup_widget(master)
        
    def _setup_widget(self, master):
        #Display Database Name
        self.label = Label(master, text = "Netflix Database")
        self.label.place(x = 200, y = 0)
        
        #Select Data to Query Label
        self._select_query_type = Label(master, text = "Select Data to Query:")
        self._select_query_type.place(x=10, y=50)
        
        #CheckButton Container for Users and Shows
        self._checkbuttonFrame1 = Frame(master)
        self._checkbuttonFrame1.place(x = 10, y = 80)
        
        #CheckButton for Users
        self.userBox = IntVar()
        self.user_checkbutton = Checkbutton(self._checkbuttonFrame1, text="Users", variable=self.userBox, command=self.displayOptionFrame)
        self.user_checkbutton.grid(row=0, column=0)
        
        #Checkbutton for Shows
        self.showBox = IntVar()
        self.shows_checkbutton = Checkbutton(self._checkbuttonFrame1, text="Shows",variable=self.showBox, command=self.displayOptionFrame)
        self.shows_checkbutton.grid(row=0, column=1)
        
        #QueryFrame
        self._queryOptionFrame = Frame(self.master)
        self._queryOptionFrame.place(x = 0, y = 110)
        
        #Display Frame
        self._data_frame = Frame(self.master)
        self._data_frame.place(x = 10, y = 350)
    
    def displayOptionFrame(self):
    
        #Remove Query Frame's Children
        for item in self._queryOptionFrame.winfo_children():
            item.grid_forget()
        
        #Remove Data Frame's Children
        for item in self._data_frame.winfo_children():
            item.grid_forget()
            
        if(self.userBox.get()):
        
            #Disable show checkbutton
            self.shows_checkbutton.config(state = DISABLED)
            
            #Select User Label
            self.select_user_label = Label(self._queryOptionFrame, text = "Select a User: ")
            self.select_user_label.grid(row = 0, column = 0)

            #User DropDown Menu
            self.menuChoice = StringVar(self._queryOptionFrame)
        
            self.userChoices = ["User 1", "User 2", "User 3", "User 4", "User 5", 
                            "User 6", "User 7", "User 8", "User 9", "User 10"]
        
            self.menuChoice.set("User 1")
                
            self._user_menu = OptionMenu(self._queryOptionFrame, self.menuChoice, *self.userChoices)
            self._user_menu.grid(row = 1, column = 0)
            
            #Recommended Shows Button 
            self.recommendShow = Button(self._queryOptionFrame, text="Recommended Shows", command = self.recommendedShowButton)
            self.recommendShow.grid(row = 1, column = 2)
    
            #Favorite Shows Button 
            self.favoriteShow = Button(self._queryOptionFrame, text="Favorite Shows", command = self.favoriteShowButton)
            self.favoriteShow.grid(row = 2, column = 2)
            
        if(self.showBox.get()):
            
            #Disable user checkbutton
            self.user_checkbutton.config(state = DISABLED)
            
            #Title Widget
            self._title_prompt = Label(self._queryOptionFrame, text = "Enter the title of the show to search for : ", justify = "left")
            self._title_prompt.grid(row = 0, column = 0, sticky=W)
        
            #Title Widget
            self._title = Label(self._queryOptionFrame, text = "Title")
            self._title.grid(row = 1, column = 0, sticky=W)
        
            #Textbox for Title - Reads User's Input
            self._title_txtbox = Entry(self._queryOptionFrame, width = 20)
            self._title_txtbox.grid(row = 1, column = 1, sticky=W)
        
            #Select Type of Show Label
            self._select_query_type = Label(self._queryOptionFrame, text = "Select Type of show:")
            self._select_query_type.grid(row = 3, column = 0, sticky=W)
        
            #CheckButton for Movies
            self.movieBox = IntVar()
            self.movies_checkbutton = Checkbutton(self._queryOptionFrame, text="Movies", variable=self.movieBox, command=self.toggleCheckButtons)
            self.movies_checkbutton.grid(row=5, column=0, sticky=W)

            #Checkbutton for TV shows
            self.tvshowBox = IntVar()
            self.tvshows_checkbutton = Checkbutton(self._queryOptionFrame, text="TV Shows",variable=self.tvshowBox, command=self.toggleCheckButtons)
            self.tvshows_checkbutton.grid(row=5, column=1, sticky=W)

            #Genre Widget
            self._genre = Label(self._queryOptionFrame, text = "Specify Genre: ")
            self._genre.grid(row = 7, column = 0, sticky=W)

            #Textbox for Genre - Reads User's Input
            self._genre_txtbox = Entry(self._queryOptionFrame, width = 10)
            self._genre_txtbox.grid(row = 7, column = 1, sticky=W)
            
            #Additional Information Widget
            self._additionalInfo = Label(self._queryOptionFrame, text = "Additional Information to display :")
            self._additionalInfo.grid(row = 8, column = 0, sticky=W)
            
            #Checkbutton for Seasons
            self.seasonBox = IntVar()
            self.season_checkbutton = Checkbutton(self._queryOptionFrame, text="Seasons",variable=self.seasonBox, command=self.toggleCheckButtons)
            self.season_checkbutton.grid(row=9, column=0, sticky=W)
     
            #CheckButton for Actors
            self.actorBox = IntVar()
            self.actor_checkbutton = Checkbutton(self._queryOptionFrame, text="Actors", variable=self.actorBox, command=self.toggleCheckButtons)
            self.actor_checkbutton.grid(row= 9, column= 1, sticky=W)

            #CheckButton for Writers
            self.writerBox = IntVar()
            self.writer_checkbutton = Checkbutton(self._queryOptionFrame, text="Writers", variable=self.writerBox, command=self.toggleCheckButtons)
            self.writer_checkbutton.grid(row=11, column= 0, sticky=W)

            #Checkbutton for Directors
            self.directorBox = IntVar()
            self.director_checkbutton = Checkbutton(self._queryOptionFrame, text="Directors",variable=self.directorBox, command=self.toggleCheckButtons)
            self.director_checkbutton.grid(row=11, column=1, sticky=W)
           
            #Search Button 
            self.search_button = Button(self._queryOptionFrame, text="Search", command=self.searchButton)
            self.search_button.grid(row = 13, column = 0, sticky=W + E, padx=60)
            
            
        #Enables disabled checkbutton 
        if(not self.showBox.get()):
            self.user_checkbutton.config(state=NORMAL)   
            
        if(not self.userBox.get()):
            self.shows_checkbutton.config(state=NORMAL)
            
    def toggleCheckButtons(self):  
    
        if(self.movieBox.get()):
            self.season_checkbutton.config(state = DISABLED)  
              
            if(self.actorBox.get()):
                self.director_checkbutton.config(state = DISABLED) 
                self.writer_checkbutton.config(state = DISABLED)  
            
            elif(self.directorBox.get()):
                self.writer_checkbutton.config(state = DISABLED) 
                self.actor_checkbutton.config(state = DISABLED) 
        
            elif(self.writerBox.get()):
                self.director_checkbutton.config(state = DISABLED) 
                self.actor_checkbutton.config(state = DISABLED) 
            else:
                self.director_checkbutton.config(state = NORMAL) 
                self.writer_checkbutton.config(state = NORMAL) 
                self.actor_checkbutton.config(state = NORMAL)
        else:
            self.season_checkbutton.config(state = NORMAL) 
            
            
            if(self.tvshowBox.get()):
                self.director_checkbutton.config(state = DISABLED)
                
                if (self.actorBox.get()): 
                    self.writer_checkbutton.config(state = DISABLED) 
                    self.season_checkbutton.config(state = DISABLED)

                elif(self.writerBox.get()):
                    self.actor_checkbutton.config(state = DISABLED)
                    self.season_checkbutton.config(state = DISABLED)

                elif(self.seasonBox.get()):
                    self.actor_checkbutton.config(state = DISABLED) 
                    self.writer_checkbutton.config(state = DISABLED) 
                else:
                    self.actor_checkbutton.config(state = NORMAL)
                    self.season_checkbutton.config(state = NORMAL)
                    self.writer_checkbutton.config(state = NORMAL) 
            else:
                self.director_checkbutton.config(state = NORMAL) 
                self.writer_checkbutton.config(state = NORMAL) 
                self.actor_checkbutton.config(state = NORMAL)
                self.season_checkbutton.config(state = NORMAL) 
                
    def recommendedShowButton(self):
        args = []
        
        getUserNumber = "%"
        getUserNumber += self.menuChoice.get()[5:7]
        getUserNumber += "%"
        
        args.append(getUserNumber)
       
        query = "SELECT shows.name, shows.description, shows.genre, shows.releaseYear FROM user, shows, recommendedshows WHERE username LIKE %s AND user.id = recommendedshows.userId AND shows.id = recommendedshows.showid"
        
        
        self.cur.execute(query, args)    
        self.displayData(self.cur.fetchall())
        
    def favoriteShowButton(self):
        args = []
        
        getUserNumber = "%"
        getUserNumber += self.menuChoice.get()[5:7]
        getUserNumber += "%"
        
        args.append(getUserNumber)
       
        query = "SELECT shows.name, shows.description, shows.genre, shows.releaseYear FROM user, shows, favoriteshows WHERE username LIKE %s AND user.id = favoriteshows.userId AND shows.id = favoriteshows.showid"
        
        
        self.cur.execute(query, args)    
        self.displayData(self.cur.fetchall())
        
    def searchButton(self):
        
        if not self.movieBox.get() and not self.tvshowBox.get():
            self.displayData([[]])
            return
        
        args = []  
        query = ""
        
        title = self._title_txtbox.get()
        title += "%"
        args.append(title)
        
        genre = self._genre_txtbox.get()
        genre += "%"
        args.append(genre)
        
        #If only Movie and Tvshow check button are selected
        if self.movieBox.get() and self.tvshowBox.get() :
            query = "SELECT name, description, genre, releaseYear FROM shows WHERE name LIKE %s AND genre LIKE %s"
        
        #If only Movie check button is selected
        elif self.movieBox.get() and not self.tvshowBox.get() and not self.actorBox.get() and not self.directorBox.get() and not self.writerBox.get():
                 query = "SELECT shows.name, shows.description, shows.genre, shows.releaseYear, movies.rating FROM shows, movies WHERE shows.name LIKE %s AND genre LIKE %s AND shows.id = movies.id"
        
        #If only TV show check button is selected
        elif self.tvshowBox.get() and not self.movieBox.get() and not self.actorBox.get() and not self.writerBox.get()  and not self.seasonBox.get() :
                query = "SELECT shows.name, shows.description, shows.genre, shows.releaseYear, tvshows.rating FROM shows, tvshows WHERE shows.name LIKE %s AND genre LIKE %s AND shows.id = tvshows.id"
        
        #If only Movie and Actor check button are selected
        elif self.movieBox.get() and self.actorBox.get():
            query = "SELECT shows.name, shows.description, shows.genre, shows.releaseYear, actors.name FROM shows, movies, starsIn, actors WHERE shows.name LIKE %s AND genre LIKE %s AND shows.id = movies.id AND starIns.showId = shows.id AND actors.id = starsIn.actorId"
        
        #If only Movie and Director check button are selected
        elif self.movieBox.get() and self.directorBox.get() :
            query = "SELECT shows.name, shows.description, shows.genre, shows.releaseYear, directors.name FROM shows, movies, directors WHERE shows.name LIKE %s AND genre LIKE %s AND shows.id = movies.id AND movies.directorId = directors.id"
        
        #If only Movie and Writer check button are selected
        elif self.movieBox.get() and self.writerBox.get() :
            query = "SELECT shows.name, shows.description, shows.genre, shows.releaseYear, writers.name FROM shows, movies, writesFor, writers WHERE shows.name LIKE %s AND genre LIKE %s AND shows.id = movies.id AND writesFor.showId = shows.id AND writers.id = writesFor.writerId"
         
        #If only TV Show and Actor check button are selected
        elif self.tvshowBox.get() and self.actorBox.get() :
            query = "SELECT shows.name, shows.description, shows.genre, shows.releaseYear, actors.name FROM shows, tvshows, starsIn, actors WHERE shows.name LIKE %s AND genre LIKE %s AND shows.id = tvshows.id AND starIns.showId = shows.id AND actors.id = starsIn.actorId"
        
        #If only TV Show and Writer check button are selected
        elif self.tvshowBox.get() and self.writerBox.get() :
            query = "SELECT shows.name, shows.description, shows.genre, shows.releaseYear, writers.name FROM shows, tvshows, writesFor, writers WHERE shows.name LIKE %s AND genre LIKE %s AND shows.id = tvshows.id AND writesFor.showId = shows.id AND writers.id = writesFor.writerId"
        
        #If only TV Show and Season check button are selected
        elif self.tvshowBox.get() and self.seasonBox.get() :
            query = "SELECT shows.name, shows.description, shows.genre, shows.releaseYear, seasons.seasonNum, seasons.releaseYear, episodes.episodeNum, episodes.releaseYear FROM shows, tvshows, seasons, episodes WHERE shows.name LIKE %s AND genre LIKE %s AND shows.id = tvshows.id AND tvshows.id = seasons.tvshowid AND seasons.id = episodes.seasonId"
       
        else:
            return
        
        self.cur.execute(query, args)
        self.displayData(self.cur.fetchall())
                
    def displayData(self, array):
        
        #Remove Frame's Children
        for item in self._data_frame.winfo_children():
            item.grid_forget()
       
        #Display Data from Database onto frame
        for index, data in enumerate(array):
            for i in range(len(data)):
                Label(self._data_frame, text = data[i], justify = "left", wraplength = 300).grid(row = index + 2, column = i)         
        
        
root = Tk()
gui = netflixInterface(root)
root.mainloop()
```
